## Soil type based crop stages dates

```python
import pandas as pd
from pathlib import Path

# --------------------------------------------------
# File paths
# --------------------------------------------------
file1 = "/Users/dulminiguruge/Documents/Seeding Date/Dates_on_crop_type1.xlsx"
#file2 = "/Users/dulminiguruge/Documents/Seeding Date/Dates_on_crop_type2.xlsx"

output_dir = Path(
    "/Users/dulminiguruge/Documents/GitHub/crop_stages_detection_2026/data/raw"
)
output_dir.mkdir(parents=True, exist_ok=True)

# --------------------------------------------------
# Reusable loader function (FINAL & SIMPLE)
# --------------------------------------------------
def load_crop_stage_excel(file_path):
    df = pd.read_excel(file_path)

    # Normalize column names
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace(r"\s+", "_", regex=True)
        .str.replace(",", "", regex=False)
    )

    # Expected schema
    expected_cols = [
        "CropType",
        "SoilType",
        "Seeding_to_Flowering",
        "Flowering_to_Filling",
        "Filling_to_Maturity",
        "Total_Days",
    ]

    missing = set(expected_cols) - set(df.columns)
    if missing:
        raise ValueError(f"Missing columns in {file_path}: {missing}")

    # Enforce numeric columns
    numeric_cols = expected_cols[2:]
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors="raise")

    # Enforce strings
    df["CropType"] = df["CropType"].astype(str)
    df["SoilType"] = df["SoilType"].astype(str)

    return df[expected_cols]

# --------------------------------------------------
# Load datasets
# --------------------------------------------------
df_set1 = load_crop_stage_excel(file1)
#df_set2 = load_crop_stage_excel(file2)

# --------------------------------------------------
# Save to Parquet
# --------------------------------------------------
df_set1.to_parquet(
    output_dir / "crop_stage_duration_by_soil_type.parquet",
    index=False
)

#df_set2.to_parquet(output_dir / "crop_stage_duration_by_soil_set2.parquet",index=False)

print("✅ Parquet files saved successfully")
```

```python
import pandas as pd

# --------------------------------------------------
# Load all Parquet files
# --------------------------------------------------
df_all = pd.concat([
    pd.read_parquet(output_dir / "crop_stage_duration_by_soil_type.parquet"),
    
], ignore_index=True)

# --------------------------------------------------
# Build dictionary
# --------------------------------------------------
crop_soil_dict = {}

for _, row in df_all.iterrows():
    soil = row["SoilType"]
    crop = row["CropType"]

    crop_soil_dict.setdefault(soil, {})

    crop_soil_dict[soil][crop] = {
        "Seeding_to_Flowering": int(row["Seeding_to_Flowering"]),
        "Flowering_to_Filling": int(row["Flowering_to_Filling"]),
        "Filling_to_Maturity": int(row["Filling_to_Maturity"]),
        "Total_Days": int(row["Total_Days"]),
    }

print("✅ Crop–soil dictionary created")
```
